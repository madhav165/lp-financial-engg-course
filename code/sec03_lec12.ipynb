```python
import pandas as pd
import numpy as np
```

```python
close_df = pd.read_csv('../data/full_close.csv', index_col='Date', parse_dates=True, dayfirst=True)
```

```python
return_df = (close_df/close_df.shift(1))-1
```

```python
return_df['APOLLOHOSP'].hist(bins=100)
```

```python
return_df.mean(axis=0)
```

```python
return_df.std(axis=0)
```

```python
log_return_df = np.log(1 + return_df)
```

```python
log_return_df.std(axis=0)
```

```python
return_df.dropna(axis=0, how='all', inplace=True)
return_df.to_csv('../data/full_return.csv', index=True)

log_return_df.dropna(axis=0, how='all', inplace=True)
log_return_df.to_csv('../data/full_log_return.csv', index=True)
```
