```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import norm
```

```python
log_return_df = pd.read_csv('../data/full_log_return.csv', index_col='Date')
```

```python
apollo = log_return_df['APOLLOHOSP']
```

```python
x_list = np.linspace(apollo.min(), apollo.max(), 100)
```

```python
y_list = norm.pdf(x_list, loc = apollo.mean(), scale = apollo.std())
```

```python
plt.hist(apollo, bins=100, density=True)
plt.plot(x_list, y_list)
```

```python
sm.qqplot(log_return_df['APOLLOHOSP'].dropna(), line='s');
```
