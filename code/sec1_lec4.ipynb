```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
```

```python
rand_arr = np.random.normal(size=1000)
```

```python
plt.plot(rand_arr)
```

```python
plt.hist(rand_arr)
```

```python
x = np.arange(0,10,0.01)
y = 1.5 * x + rand_arr
```

```python
model = LinearRegression()
model.fit(x.reshape(-1,1), y)
```

```python
m = model.coef_[0]
```

```python
y2 = m * x
```

```python
plt.scatter(x, y)
plt.plot(x, y2, color = 'orange')
```

```python
plt.plot(np.cumsum(rand_arr))
```

```python
mv_rand_arr = np.random.multivariate_normal([0,0],[[1,0],[0,1]], size=(1000))
```

```python
plt.scatter(mv_rand_arr[:,0], mv_rand_arr[:,1])
```

```python
np.mean(mv_rand_arr, axis=0)
```

```python
n = mv_rand_arr.shape[0]
u1 = sum(mv_rand_arr[:,0])/n
u2 = sum(mv_rand_arr[:,1])/n
print(np.array([u1, u2]))
```

```python
np.cov(mv_rand_arr, rowvar=False)
```

```python
c11 = sum((mv_rand_arr[:,0] - u1)**2)/(n-1)
c12 = sum((mv_rand_arr[:,0] - u1)*(mv_rand_arr[:,1] - u2))/(n-1)
c21 = sum((mv_rand_arr[:,1] - u2)*(mv_rand_arr[:,0] - u1))/(n-1)
c22 = sum((mv_rand_arr[:,1] - u2)**2)/(n-1)
print(np.array([[c11, c12], [c21, c22]]))
```
