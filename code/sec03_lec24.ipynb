```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
log_return_df = pd.read_csv('../data/full_log_return.csv', index_col='Date', parse_dates=True, dayfirst=True)
```

```python
apollo = log_return_df['APOLLOHOSP']
```

```python
n = len(apollo)
```

```python
mean = np.sum(apollo)/n
mean
```

```python
var = np.sum((apollo - mean)**2)/n
std = np.sqrt(var)
std
```

```python
skew = np.sum(((apollo - mean)/std)**3)/n
skew
```

```python
kurtosis = np.sum(((apollo - mean)/std)**4)/n - 3
kurtosis
```

```python
conf_min = mean - (1.96*std/np.sqrt(n))
conf_max = mean + (1.96*std/np.sqrt(n))
```

```python
plt.figure()
apollo.hist(bins=100, density=True)
plt.xlim(apollo.min() - 0.001, apollo.max() + 0.001)
plt.axvline(mean, label='mean', color='red')
plt.axvline(conf_min, label='conf_min', color='green')
plt.axvline(conf_max, label='conf_max', color='green')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.xlim(conf_min - 0.001, conf_max + 0.001)
plt.axvline(0, label='zero', color='blue')
plt.axvline(mean, label='mean', color='red')
plt.axvline(conf_min, label='conf_min', color='green')
plt.axvline(conf_max, label='conf_max', color='green')
plt.legend()
plt.show()
```
